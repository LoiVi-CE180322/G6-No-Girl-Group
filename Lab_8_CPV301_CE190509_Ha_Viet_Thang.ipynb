```python
import cv2

input_path = '/content/video.mp4'
output_path = '/content/output_tracked_video.mp4'

cap = cv2.VideoCapture(input_path)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

initial_box = (300, 400, 200, 200)

tracker = cv2.TrackerCSRT_create()
ret, frame = cap.read()
if not ret:
    cap.release()
    out.release()

tracker.init(frame, initial_box)
trajectory_points = []

while True:
    ret, frame = cap.read()
    if not ret:
        break

    success, box = tracker.update(frame)

    if success:
        x, y, w, h = [int(v) for v in box]
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 255, 255), 2)
        center_x, center_y = x + w // 2, y + h // 2
        trajectory_points.append((center_x, center_y))
    else:
        tracker = cv2.TrackerCSRT_create()
        tracker.init(frame, initial_box)
        x, y, w, h = initial_box
        center_x, center_y = x + w // 2, y + h // 2
        trajectory_points.append((center_x, center_y))

    for i in range(1, len(trajectory_points)):
        cv2.line(frame, trajectory_points[i - 1], trajectory_points[i], (0, 0, 255), 2)

    out.write(frame)

cap.release()
out.release()
print(f"Video output với đường hành trình giữ nguyên được lưu tại: {output_path}")
```
