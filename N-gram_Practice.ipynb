```python
import itertools
from collections import defaultdict, Counter
import ipywidgets as widgets
from IPython.display import display, clear_output

# Định nghĩa corpus
van_ban_mau = """The girl bought a chocolate
The boy ate the chocolate
The girl bought a toy
The girl played with the toy"""


# Tiền xử lý văn bản
def tien_xu_ly_van_ban(van_ban):
    van_ban = van_ban.lower()
    tu = van_ban.split()
    return tu

danh_sach_tu = tien_xu_ly_van_ban(van_ban_mau)

# Xây dựng mô hình ba-gram
ba_gram = list(zip(danh_sach_tu[:-2], danh_sach_tu[1:-1], danh_sach_tu[2:]))
dem_so = Counter(ba_gram)
ngu_canh_dem_so = defaultdict(Counter)

for (tu1, tu2, tu3), so_luong in dem_so.items():
    ngu_canh_dem_so[(tu1, tu2)][tu3] = so_luong


def goi_y_tu_tiep_theo(ngu_canh, top_k=3):
    cac_tu = ngu_canh.lower().split()
    if len(cac_tu) < 2:
        return ["Không có gợi ý nào"]

    cap_ngu_canh = (cac_tu[-2], cac_tu[-1])
    if cap_ngu_canh in ngu_canh_dem_so:
        goi_y = ngu_canh_dem_so[cap_ngu_canh].most_common(top_k)
        return [tu for tu, _ in goi_y]
    else:
        return ["Không có gợi ý nào"]


truong_nhap = widgets.Text(placeholder="Nhập văn bản của bạn...")
cac_nut_goi_y = [widgets.Button(description="") for _ in range(3)]
khu_vuc_xuat = widgets.Output()

def cap_nhat_goi_y(change):
    ngu_canh = truong_nhap.value.strip()
    goi_y = goi_y_tu_tiep_theo(ngu_canh)

    for i, nut in enumerate(cac_nut_goi_y):
        if i < len(goi_y):
            nut.description = goi_y[i]
        else:
            nut.description = ""

def khi_nut_duoc_bam(nut):
    truong_nhap.value = truong_nhap.value.strip() + " " + nut.description
    cap_nhat_goi_y(None)

for nut in cac_nut_goi_y:
    nut.on_click(khi_nut_duoc_bam)

truong_nhap.observe(cap_nhat_goi_y, names="value")

giao_dien = widgets.VBox([truong_nhap] + cac_nut_goi_y)
display(giao_dien)
```
