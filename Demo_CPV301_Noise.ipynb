Mean Filter (3x3 kernel blur)

```python
import cv2
from google.colab.patches import cv2_imshow

# Load image in grayscale
noisy_image = cv2.imread('/content/sample_data/noisy_image.jpg')

# Apply Mean Filter (3x3 kernel)
mean_filtered_image = cv2.blur(noisy_image, (3, 3))

# Display the original image
cv2_imshow(noisy_image)
# Display the de-noised image using 3x3 kernel
cv2_imshow(mean_filtered_image)

# Save the image after de-noise
cv2.imwrite('mean_filtered_image.jpg', mean_filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Median Filter (3x3 kernel blur)

```python
import cv2
from google.colab.patches import cv2_imshow

# Load the image
noisy_image = cv2.imread('/content/sample_data/noisy_image.jpg')

# Apply Median Filter (kernel size must be odd, e.g., 3, 5, 7)
median_filtered_image = cv2.medianBlur(noisy_image, 2)


# Display the original noisy image
cv2_imshow(noisy_image)
# Display the de-noised image using the Median Filter
cv2_imshow(median_filtered_image)

# Save the de-noised image
cv2.imwrite('median_filtered_image.jpg', median_filtered_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
