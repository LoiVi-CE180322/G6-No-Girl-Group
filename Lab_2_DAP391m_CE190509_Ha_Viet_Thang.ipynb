```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the dataset
file_path = '/content/lab2 Topic_Survey_Assignment.csv'
df = pd.read_csv(file_path)

# Total number of respondents
total_respondents = 2233

# Calculate percentage for each category
df['Very interested (%)'] = (df['Very interested'] / total_respondents) * 100
df['Somewhat interested (%)'] = (df['Somewhat interested'] / total_respondents) * 100
df['Not interested (%)'] = (df['Not interested'] / total_respondents) * 100

# Sort topics based on 'Very interested (%)'
df = df.sort_values(by='Very interested (%)', ascending=False)

# Set up the positions for each bar
categories = df['Unnamed: 0']
bar_width = 0.25
r1 = np.arange(len(categories))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create a grouped bar plot
plt.figure(figsize=(15, 6))

# Plot the bars
bars1 = plt.bar(r1, df['Very interested (%)'], color='#5cb85c', width=bar_width, edgecolor='grey', label='Very interested')
bars2 = plt.bar(r2, df['Somewhat interested (%)'], color='#5bc0de', width=bar_width, edgecolor='grey', label='Somewhat interested')
bars3 = plt.bar(r3, df['Not interested (%)'], color='#d9534f', width=bar_width, edgecolor='grey', label='Not interested')

# Add the percentage values on top of each bar
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.2f}%', ha='center', va='bottom')

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.2f}%', ha='center', va='bottom')

for bar in bars3:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.2f}%', ha='center', va='bottom')

plt.ylim(0, 85)  # Extend the Y-axis limit slightly above the highest bar

# Add labels and titles
plt.xlabel('Data Science Topics', fontsize=12)
plt.ylabel('Percentage of Respondents', fontsize=12)
plt.title("Percentage of Respondents' Interest In Data Science Areas", fontsize=16)
plt.xticks([r + bar_width for r in range(len(categories))], categories, rotation=45, ha='right', fontsize=12)

# Add the legend
plt.legend(loc='upper right', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```
