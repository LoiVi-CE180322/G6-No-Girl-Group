```python
import ibm_db
import os
import pandas as pd
from flask import Flask, render_template, request, redirect, url_for
from pyngrok import ngrok

# Set your IBM Db2 credentials
dsn_hostname = '815fa4db-dc03-4c70-869a-a9cc13f33084.bs2io90l08kqb1od8lcg.databases.appdomain.cloud'
dsn_uid = 'drl64387'
dsn_pwd = 'nCd2UC0r6DXLjRrz'
dsn_port = '30367'
dsn_database = 'bludb'
dsn_driver = '{IBM DB2 ODBC DRIVER}'
dsn_protocol = 'TCPIP'
dsn_security = 'SSL'

# Create the DSN connection string
dsn = (
    "DRIVER={0};"
    "DATABASE={1};"
    "HOSTNAME={2};"
    "PORT={3};"
    "PROTOCOL={4};"
    "UID={5};"
    "PWD={6};"
    "SECURITY={7};").format(
    dsn_driver, dsn_database, dsn_hostname, dsn_port, dsn_protocol, dsn_uid, dsn_pwd, dsn_security)

# Initialize the Flask app
app = Flask(__name__)

# Function to get data from the database
def get_data():
    try:
        # Connect to the database
        conn = ibm_db.connect(dsn, "", "")
        # Fetch data
        sql = "SELECT * FROM STUDENT_LIST"
        stmt = ibm_db.exec_immediate(conn, sql)
        students = []
        result = ibm_db.fetch_assoc(stmt)
        while result:
            students.append(result)
            result = ibm_db.fetch_assoc(stmt)
        return students
    except Exception as e:
        print("Error fetching data:", str(e))
        return []
    finally:
        if 'conn' in locals() and conn:
            ibm_db.close(conn)

# Define the route to render the HTML template with data
@app.route('/')
def home():
    students = get_data()  # Get the latest data
    return render_template('home.html', students=students)

# Route to handle adding a new student
@app.route('/add', methods=['POST'])
def add_student():
    # Get form data
    student_name = request.form['Tên Sinh Viên']
    date_of_birth = request.form['Ngày Sinh']
    student_id = request.form['Mã Số']
    major = request.form['Chuyên Ngành']

    try:
        # Connect to the database
        conn = ibm_db.connect(dsn, "", "")
        # Prepare and execute the INSERT statement
        sql = 'INSERT INTO STUDENT_LIST ("Tên Sinh Viên", "Ngày Sinh", "Mã Số", "Chuyên Ngành") VALUES (?, ?, ?, ?)'
        stmt = ibm_db.prepare(conn, sql)
        ibm_db.bind_param(stmt, 1, student_name)
        ibm_db.bind_param(stmt, 2, date_of_birth)
        ibm_db.bind_param(stmt, 3, student_id)
        ibm_db.bind_param(stmt, 4, major)
        ibm_db.execute(stmt)
        print(f"Student {student_name} added successfully.")
    except Exception as e:
        print("Error adding student:", str(e))
    finally:
        if 'conn' in locals() and conn:
            ibm_db.close(conn)
    return redirect(url_for('home'))

# Route to handle deleting a student
@app.route('/delete', methods=['POST'])
def delete_student():
    student_id = request.form['Mã Số']
    try:
        # Connect to the database
        conn = ibm_db.connect(dsn, "", "")
        # Prepare and execute the DELETE statement
        sql = 'DELETE FROM STUDENT_LIST WHERE "Mã Số" = ?'
        stmt = ibm_db.prepare(conn, sql)
        ibm_db.bind_param(stmt, 1, student_id)
        ibm_db.execute(stmt)
        print(f"Student with Mã Số {student_id} deleted successfully.")
    except Exception as e:
        print("Error deleting student:", str(e))
    finally:
        if 'conn' in locals() and conn:
            ibm_db.close(conn)
    return redirect(url_for('home'))

# Function to run the Flask app with ngrok
def run_app():
    # Terminate any existing ngrok tunnels
    ngrok.kill()
    # Start ngrok tunnel
    public_url = ngrok.connect(5000).public_url
    print(f" * Ngrok Tunnel URL: {public_url}")
    # Run the Flask app
    app.run(port=5000)

# Start the app
run_app()
```
