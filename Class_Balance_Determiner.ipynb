```python
import os
from collections import Counter

# Function to analyze class distributions

def analyze_class_distribution(folder_path):
    class_counts = Counter()

    # Define valid image extensions
    image_extensions = [".png", ".jpg", ".jpeg", ".bmp", ".gif"]

    # Walk through the folder structure, excluding 'test' set
    for root, dirs, files in os.walk(folder_path):
        if 'test' in root.lower():
            continue
        for file in files:
            if file.lower().endswith(tuple(image_extensions)):
                # Assuming the folder name represents the class
                class_name = os.path.basename(root)
                class_counts[class_name] += 1

    # Sort class counts from lowest to highest
    sorted_class_counts = sorted(class_counts.items(), key=lambda x: x[1])

    # Print class distribution
    print("Image class distribution:")
    for class_name, count in sorted_class_counts:
        print(f"{class_name}: {count}")

    # Optional: Visualize class distribution as a bar chart
    try:
        import matplotlib.pyplot as plt
        classes, counts = zip(*sorted_class_counts)
        plt.bar(classes, counts)
        plt.xlabel('Classes')
        plt.ylabel('Number of Images')
        plt.title('Class Distribution')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
    except ImportError:
        print("Matplotlib not installed. Install it to visualize class distribution.")

# Mount Google Drive to access files (if required)
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# Specify the folder path
main_folder_path = '/content/drive/My Drive/DAC_Data'  # Replace with your main folder path

# Analyze class distribution
analyze_class_distribution(main_folder_path)
```
